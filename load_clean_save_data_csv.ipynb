```python
import pandas as pd
```

```python
# load dataset of cultural object 
cultural_objects_df = pd.read_csv('Seattle_Cultural_Space_Inventory.csv')
```

```python
cultural_objects_df.head(3)
```

```python
# cleanse data - exclude from result set places without name or exact location
cultural_objects_clean_df = cultural_objects_df[(~cultural_objects_df.Name.isnull()) & (~cultural_objects_df.Latitude.isnull()) & (~cultural_objects_df.Longitude.isnull())]
```

```python
cultural_objects_clean_df.head(3)
```

```python
# Let's see if number of rows changed for cultural places data set 
len(cultural_objects_df) - len(cultural_objects_clean_df)
```

```python
# load 911 fire calls dataset
fire_calls_df = pd.read_csv('Seattle_Real_Time_Fire_911_Calls.csv')
```

```python
fire_calls_df.head(3)
```

```python
fire_calls_clean_df = fire_calls_df[(~fire_calls_df.Type.isnull()) & (~fire_calls_df.Datetime.isnull()) & (~fire_calls_df.Latitude.isnull()) & (~fire_calls_df.Longitude.isnull())]
```

```python
# Let's see if number of rows changed for fire calls data set 
len(fire_calls_df) - len(cultural_objects_clean_df)
```

```python
# Drop unnecessary columns 
cultural_objects_clean_df = cultural_objects_clean_df.drop(['Phone', 'URL', 'Square Feet Total', 'Neighborhood',
       'Organization Type', 'Dominant Discipline', 'Year of Occupation',
       'Rent vs Own', 'Age of Current Building', 'Length of Lease (Date)',
       ' Year Organization Founded', 'Number of Past Facilities',
       'Stages and Theaters', 'Stage & Theater Seats ', 'Gallery Space',
       'Gallery Square Feet', 'ADA Compliant', 'Available Parking',
       'Street Presence', 'Rental Space ', 'Alcohol Sales',
       'Organization Mission', 'Funded by A&C', 'Funded by 4Culture',
       'Stability Index (5=very stable, 1=very uncertain)',
       'Control Index (5=very in control, 1 = very out of control) ',
       'Constituency over 50% one race', 'Specific Demographics and Community',
       'Organization Leadership', 'Organization Artists', 'Closed Date',
       'Closed?', 'Location'], axis=1)
```

```python
fire_calls_clean_df = fire_calls_clean_df.drop(['Report Location'], axis=1)
```

```python
#now let's add technical fields for future analysis
# Range Latitude Up - lalitide of cultiral object + 1 km 
# Range Latitude Down - lalitide of cultiral object - 1 km 
# Range Longitude Up - Longitude of cultiral object + 1 km 
# Range Longitude Down - Longitude of cultiral object - 1 km 
from numpy import cos, pi
```

```python
# Earth radius
r_earth = 6378
```

```python
# "Range" - we will analyze fire calls in 1 km box around cultural building
range_km = 1
```

```python
cultural_objects_clean_df['Range Latitude Up'] = cultural_objects_clean_df['Latitude'] + (range_km / r_earth) * (180 / pi)
```

```python
cultural_objects_clean_df['Range Latitude Down'] = cultural_objects_clean_df['Latitude'] - (range_km / r_earth) * (180 / pi)
```

```python
cultural_objects_clean_df.head(3)
```

```python
cultural_objects_clean_df.dtypes
```

```python
cultural_objects_clean_df['Range Longitude Up'] = cultural_objects_clean_df['Longitude'] + (range_km / r_earth) * (180 / pi) / cos(cultural_objects_clean_df['Latitude'] * pi/180)
```

```python
cultural_objects_clean_df['Range Longitude Down'] = cultural_objects_clean_df['Longitude'] - (range_km / r_earth) * (180 / pi) / cos(cultural_objects_clean_df['Latitude'] * pi/180)
```
